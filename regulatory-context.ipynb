# Regulatory frameworks in Credit Risk 

## Why do banks need to model risk? 

Modelling risk is inherently important to a bank as it is something vested in its definition as an institution
    -\textit{Banks act as intermediaries between borrowers and lenders. If the bank lends out money to a person, and the person defaults, the bank would not be able to pay the lender back.}

This perhaps is the crux of the problem. Therefore, banks need to do SOMETHING in the case the person they lent the money to, defaults. 

What the bank can do here is to hold out capital against the loan it gives out. For instance, if it is going out a loan to a person the bank (somehow-- and we will get to this in a bit) deems risky, then the bank must have the same (or almost the same) amount of cash with it, in case the person does default. The more risky the party whom the bank is lending money, the more capital they will have to hold against it. 

Since, people's hard earned money was involved. This needed to be regulated, that is, the need to demarcate rigorously how much capital had to be held for a given level of risk profile. For that, we had the Basel Accords.

Basel-I was somewhat crude. The banks were just asked to hold 8% capital against the risk weighted assets; all corporate loans were equal in terms of risk the bank had taken on. Basel-II then revisited this in 2004, including the necessity to actually compute the risk the bank takes on for each loan it gives out. They then could either have a standardized, or an internal rating based system to decide how much capital they would have to hold against a loan they had to give out. 

However, over time, as financial products and their derivatives became more sophisticated, systemic risks grew. Perhaps, institutions were not aware of how to compute the risk they were taking on in trading these risky derivative products. One of these products, was a mortgage backed security (MBS). A derivative security that derived its value from underlying mortgages. As trading volumes of these MBS increased, more and more MBS were constructed with underlying mortagages being subprime (risky). Soon the MBS trading grew extensively high, and the mortages underlying these MBS started to default. It turned out many ratings agencies (Moody's, S&P) were also rating risky loans as less risky. The whole system was taking on a huge collective risk, and then, in 2008, like blocks of lego as people started defaulting and one of the banks failed, it all came crumbling down. People lost confidence in the banks. Wall street was in turmoil. The rest is history. 

Basel-III was then introduced as a tighter regulatory framework for banks post-2008-- to prevent another economic breakdown happening. 

The major governing idea for any regulatory framework is what a bank's expected loss is on a loan it gives out, this is given as a function of probability of default (PD), the exposure at default (EAD), and the loss given default (LGD)

$$EL = PD\times EAD \times LGD $$

This expected loss is, as provisioned for in the income statement, but the market fluctuates, and the tail risks can be significantly different from what is expected. The capital allocated as a safety net, thus is shields from such unexpected losses. 

The question still remains, if said losses are unexpected how do banks go about computing how much capital to hold against a given loan. Under Basel-II, banks had a choice of either a standardized rating system (per loan type), or an internal-ratings based system (IRB). In the Standardized approach, for instance, all mortgage loans given out had a risk weight of 35%, and $8\%$ worth of capital had to be secured against it, however, the 35% rating was ubiquitous, regardless of the risk profile of the borrower. One might think that this would make banks less prone to suffering losses, however, the other side of this double-edged sword meant that they had to, often, secure more capital than needed for 'safe' borrowers. Banks thus favour an Internal Based Ratings system that allows them to develop models, like in the other notebook, that determines the risk profile of a borrower and then judiciously assigns how much capital to hold against it. This allows bank to maintain the subtle balance between being safe from defaults, while also maximizing profits that come from lending. There are two variants of IRB. In the **Foundation IRB (F-IRB)** approach, 
banks estimate only the Probability of Default (PD), while the regulator 
prescribes values for Loss Given Default (LGD) and Exposure at Default (EAD). 
The **Advanced IRB (A-IRB)** approach grants banks full autonomy — they 
estimate all three components internally, allowing for the most risk-sensitive 
capital calculations.

It must be said that these internal based ratings, are not arbitrary. They are carefully vetted and must be tested internally for at least 3 years, before a financial regulatory board approves of it for use on real capital. In Canada, this is the Office of the Superindendent of Financial Institutions (OSFI).


## How does our model fit into the Internal Ratings Based system?

The model we developed in the credit-risk.ipynb notebook estimates probability of default within 12 months. In this the model finds itself within the **F-IRB** framework. The probability of default in the next 12 months is the standard definition in the Basel-II framework. 

We develop a logistic regression model and run it on Home loan data downloaded from Kaggle. While sophisticated and perhaps, even more accurate, deep learning models exist, the reason for choosing logistic regression is three-fold:

1. It is end-to-end explainable to regulators. Banks can decompose each decision made by the model to find why it says what it says. This is in stark contrast to the black-boxy nature of neural networks. 
2. The logistic regression is easily transferable to a points based system. We used results from our model and converted to a FICO style scorecard. 
3. Logistic regression is stable under post-facto emergence of information, for example, a changing of data-standards, or the availability of a new feature in the dataset. 

To prepare features for our model, we use Weight-of-evidence (WoE) encoding to deal with the numerical and categorical data in the dataset, and then capture the Information Value (IV) for each feature. This is the industry standard for IRB models because it handles all data on a consistent scale, captures non-linear relationships through binning, and produces a model that is fully transparent — every feature's contribution to the final score can be traced and explained.

The results from our exercise are illuminating, especially since our model was built on real data. Our validation metrics report a Gini coefficient of 0.47, a KS statistic of 0.35, and a PSI of 0.0003 — all within regulatory benchmarks for retail credit models.

While a fruitful exercise in learning how banks model credit risk, our exercise was limited. While this model is of an academic nature, deployment in a production banking environment would require three years of internal validation before receiving approval for regulatory capital purposes from OSFI. A further exploration might involve modelling all three EAD, LGD, and PD. Moreover, a good test would be to add in a machine-learning framework for comparison, XGBoost would be a good starting point.
